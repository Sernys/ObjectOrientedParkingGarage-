# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
# Parking Garage Project
# Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 
class ParkingGarage(): 
    def __init__(self):
        self.available_tickets = 5
        #You will need a few attributes as well:
        self.ticket = ["A", "B", "C", "D", "E"]# - tickets -> list
        self.parking_space = []# - parkingSpaces -> list
        self.current_ticket = {'Paid':False}# - currentTicket -> dictionary   
 

    def takeTicket(self):
        if self.available_tickets > 0:
            print(f'Available space:{self.ticket}')
            pick_space = input('Please pick a ticket from the available space:')
            if pick_space.lower() == "a" or "b" or "c" or "d" or "e": 
                    self.ticket.remove(pick_space.upper())
                    self.parking_space.append(pick_space.upper())
                    print(f'Spot that you have choose is:[{pick_space.upper()}]')                    
                    self.available_tickets -= 1 #This should decrease the amount of tickets available by 1
            elif self.available_tickets <= 0 :
                print(f'{self.available_tickets} are availavle in out garage,come back another time.')
    
    def payForParking(self):
        if self.available_tickets > 0 :
            payment = input('Please enter your $10 payment now:')#Display an input that waits for an amount from the user and store it in a variable
            if payment == '10':
                print('Your Ticket has been paid and they have 15 minutes to leave the parking!')
                print('________________________________________________________________________\n')
                self.current_ticket['Paid'] = True
            else:
                return 'Please make your payment before you leave the parking!!'
        elif self.available_tickets <= 0:
            return f'The number of parking space we have {self.available_tickets}, come back another time.'
        
    def leaveGarage(self):
        if self.current_ticket['Paid'] == False:
            self.payForParking()#- If the ticket has not been paid, display an input prompt for payment
        if self.current_ticket['Paid'] == True: #This should update the "currentTicket" dictionary key "paid" to True
            print(f'These are the parking that are taken:{self.parking_space}')
            return_space = input("Please enter the ticket you have to returned is: ")
            self.parking_space.remove(return_space.upper())
            self.ticket.append(return_space.upper())
        print("Ticket has been paid you have 15 min to leave the parking!\n")
        print("Thank you, have a nice day hope we see you again in our parking!")
        print('________________________________________________________________\n')
        self.available_tickets += 1 #Update tickets list to increase by 1 
```

```python
GarageAccess = ParkingGarage()
while True:
    print("Welcome,Would you like to use our Parking?\nSelect from: \n-[Y] for yes\n-[N] for no\n-[L] for leaving the Parking\n-[Q] to quit")
    take_ticket = input("==>")
    if take_ticket.lower() == "y":
        GarageAccess.takeTicket()
        GarageAccess.payForParking()
    elif take_ticket.lower() == 'n':
        print("Have a Nice Day, Come back soon,GoodBye!\n")      
    elif take_ticket.lower() == 'l':
        GarageAccess.leaveGarage()        
    elif take_ticket.lower() == 'q':
        print("Have a nice day, GoodBye!")
        print("_________________________\n")
        break
    else:
        print("ERROR, please pick from the list!\n")
```
